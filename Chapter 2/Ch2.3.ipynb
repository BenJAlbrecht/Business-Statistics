## A Frequency Distribution for a Numerical Variable
Frequency distributions for categorical variables were simple, we just needed to count each category. Frequency distributions for numerical variables require the creation of intervals (sometimes called classes). We need to make decisions about the width and number of intervals. These guidelines will help us with that:
* *Intervals are mutually exclusive.* Suppose the first two intervals of a distribution are $0 < x \leq 1$ and $1 < x \leq 2$, then if $x=1$ it falls into interval 1. Intervals don't overlap, and each observation falls into only one.
* *#-of intervals usually ranges 5-20.* Large datasets will have more intervals than small datasets (in general). Note if we have too many intervals, then we can't depict the distribution together. However, if we have too few intervals then we'll lose detail.
* *Intervals are exhaustive.* The entire sample (or population) is covered by the intervals.
* *Interval limits are easy to interpret* We want intervals defined by numbers like $-10, 10$ not $-8, 12$.
For the starting point of each interval we often use the formula:
$$
\frac{Maximum - Minimum}{Number  of  Intervals}
$$

## A Histogram
We use a **histogram** to graphically depict a frequency distribution for a numerical variable. A histogram allows us to quickly see where observations cluster, as well as the spread and shape of the variable. In general a histogram is symmetric or skewed.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data
PG_Sales = pd.read_csv('PG_Sales.csv', index_col=False)
PG_Sales.head(5)
```

```python
# Graph.
fig, ax = plt.subplots()

ax.hist(PG_Sales['Price'], bins=6)
plt.show()
```

## A Polygon
A **polygon** provides another way to graph a frequency distribution. There are great ways to do this in Excel. But for using Python we can plot a continuous frequency distribution with Seaborn's kernel density estimation (KDE) plot.

```python
import seaborn as sns

sns.kdeplot(data=PG_Sales, x='Price')
plt.show()
```

## An Ogive
An **ogive** is also useful for visually conveying information. This is analagous to the CDF of a distribution. Again, we can use seaborn to plot the ECDF.

```python
sns.ecdfplot(data=PG_Sales, x='Price')
plt.show()
```
